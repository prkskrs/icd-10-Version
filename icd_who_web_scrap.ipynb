```python

def lsttotable(lst):
    ids=[]
    description=[]
    for i in range(0,len(lst)):
        x=""
        cnt=0
        for j in range(0,len(lst[i])):
            x+=lst[i][j]
            if(lst[i][j]==" " and cnt<1):
                cnt+=1
                ids.append(x)
                description.append(lst[i][j+1:])
            else:
                continue
    print(len(ids),len(description))
    return [ids,description]
```
